##**ไอคอนรูป"ตะกร้า"(Shopping Card/Basket)**
เปรียบเสมือนตะกร้าสินค้าในห้างสรรพสินค้าจริง โดยทำหน้าที่เป็นเหมือนการพักสินค้าชั่วคราว และเป็นจุดตรวจสอบและคำนวณเงิน และที่สำคัญคือการใช้เปรียบเทียบราคาสินค้า

##ข้อมูลที่สนใจเกี่ยวกับพฤติกรรมการกดเพิ่มสินค้าลงตะกร้าของลูกค้า
เช่น :
* ชนิดสินค้าและราคาสินค้าในตะกร้า
* วันและเวลาที่กดเพิ่มสินค้า (เช่น กดตอนกลางวัน เที่ยงวัน หรือกดตอนกลางคืน)
* ผลลัพธ์สุดท้าย (กดซื้อทันที ดองทิ้งไว้หลายวันแล้วค่อยซื้อ หรือลบทิ้ง)

##ไอเดียการนำข้อมูลไปใช้ทำโปรเจคต์ Data Science
โปรเจคต์ **"สำรวจพฤติกรรม : ลูกค้ากดสินค้าใส่ตะกร้าไปเพื่ออะไร"**
โดยนำข้อมูลมาพลอตกราฟและวิเคราะห์สถิติ เพื่อจัดกลุ่มคำตอบ ว่าลูกค้ากดใส่ตะกร้าด้วยเหตุผลอะไร เช่น :
1. **กดเพื่อซื้อจริง:** พวกที่กดกดปุ๊บจ่ายปั๊บ ทันที
2. **กดเพื่อเปรียบเทียบ:** พวกที่กดสินค้าชนิดเดียวกันหลาย ๆ ร้าน แล้วมาเลือกตัวที่คุ้มที่สุดทีหลัง
3. **กดเพื่อบันทึกไว้ก่อน:** พวกที่ชอบกดทิ้งไว้ตอนกลางคืน แล้วค่อยกลับมาดูหรือรอเงินเดือนออก

##ประโยชน์ที่ได้รับจากโปรเจกต์
1. **เข้าใจพฤติกรรมลูกค้าที่แท้จริง:** ร้านค้าจะรู้ Insight ว่าลูกค้าไม่ได้กดใส่ตะกร้าเพื่อจะซื้อทันทีเสมอไป บางคนแค่กดเซฟไว้เป็น Wishlist หรือกดเพื่อรอเปรียบเทียบราคา
2. **ช่วยเพิ่มยอดขาย (Boost Sales):** เมื่อรู้ว่าลูกค้ากดใส่ตะกร้าเพื่อ "รอเงินเดือนออก" หรือ "รอส่วนลด" ร้านค้าสามารถตั้งเวลาส่งข้อเสนอพิเศษหรือโค้ดลดราคาไปกระตุ้นลูกค้าได้ถูกจังหวะ
3. **ลดต้นทุนการทำการตลาด:** ไม่จำเป็นต้องแจกโค้ดส่วนลดให้ลูกค้าทุกคนจนร้านขาดทุน แต่เลือกแจกเฉพาะกลุ่มที่กำลังลังเลและเปรียบเทียบราคาอยู่เพื่อปิดการขาย
4. **ช่วยบริหารจัดการสต็อกสินค้า:** สินค้าชิ้นไหนที่มีคนกดใส่ตะกร้าค้างทิ้งไว้เยอะ ๆ แปลว่ามีความต้องการสูง ร้านค้าจะได้รีบเตรียมสต็อกหรือสั่งมาเพิ่มล่วงหน้า ไม่ปล่อยให้ของขาดตลาด
5. **ปรับปรุงระบบหน้าร้านให้ซื้อง่ายขึ้น:** ถ้าพบว่าลูกค้ากดใส่ตะกร้าเยอะมากแต่ไม่ยอมซื้อ อาจเป็นเพราะขั้นตอนการจ่ายเงินยุ่งยากเกินไป ร้านค้าจะได้นำไปปรับปรุงระบบ UI/UX ให้ใช้งานง่ายขึ้น
